```python
# ============================================================
# Handling Missing Values in 'country'
#
# Strategy:
# 1. For each row with a missing country:
#    a) Check if the director is available.
#       - Find all other movies/shows by the same director.
#       - Assign the most frequently occurring country (mode).
#
#    b) If the director cannot determine the country,
#       use the cast information.
#       - Split the cast into individual actors.
#       - Search for all titles containing each actor.
#       - Collect the countries of those titles.
#       - Assign the most frequently occurring country (mode).
#
# 2. If neither the director nor the cast provides enough
#    information, fill the remaining missing values with
#    "Unknown".
#
# This approach performs domain-based imputation instead of
# simply filling all missing values with the overall mode,
# making the data cleaning process more meaningful.
# ============================================================
```

```python
import pandas as pd

# Get indices where country is missing
missing_idx = df[df['country'].isna()].index

for idx in missing_idx:

    # -----------------------------
    # 1. Try filling using Director
    # -----------------------------
    director = df.loc[idx, 'director']

    if pd.notna(director):

        temp = df[
            (df['director'] == director) &
            (df['country'].notna())
        ]

        if not temp.empty:
            df.loc[idx, 'country'] = temp['country'].mode()[0]
            continue      # Country filled, move to next row

    # -------------------------
    # 2. Try filling using Cast
    # -------------------------
    cast = df.loc[idx, 'cast']

    if pd.notna(cast):

        actors = [actor.strip() for actor in cast.split(',')]

        countries = []

        for actor in actors:

            temp = df[
                (df['cast'].str.contains(actor, regex=False, na=False)) &
                (df['country'].notna())
            ]

            countries.extend(temp['country'].tolist())

        if countries:
            df.loc[idx, 'country'] = pd.Series(countries).mode()[0]
            continue

# -----------------------
# 3. Fill remaining rows
# -----------------------
df['country'] = df['country'].fillna('Unknown')
```
